# 🌀 Fibonacci Sequence Generator

This notebook generates the Fibonacci sequence up to a user-defined number of terms using functions, demonstrating recursive-like linear generation and basic control structures.


```python
def fib(n):
    a=0
    b=1
    if n == 0:
        print ("Invalid input")
    elif n == 1:
        print (a)
    else:
        print (a)
        print (b)
    for i in range (2,n) :
        c=a+b
        a=b
        b=c
    print (c)
num = int (input ("enter the value of n") )
fib (num)
```
